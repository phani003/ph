```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#step1
#requesting from website

responce=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')

print(responce)

```

```python
#step2
#read dato once cheack it is available or not

soup=BeautifulSoup(responce.content,'html.parser')

#print(soup)


#prettify is useful to some changes in data for readble 

#print(soup.prettify)
```

```python
#step 3 

#collection data from web site it is available it ustrared format 

#collecting movie name from website 

names=soup.find_all('td', class_="titleColumn")
     
#print(names)

#create list for storing the data
name=[]

for i in range(0,len(names)):
    name.append(names[i].get_text())
```

```python
#rating colunm

ratings=soup.find_all('td', class_="ratingColumn imdbRating")
rate=[]
for i in range (0,len(ratings)):
    rate.append(ratings[i].get_text())

```

```python
#year
years=soup.find_all('span',class_="secondaryInfo")
year=[]
for i in range (0,len(years)):
    year.append(years[i].get_text())

```

```python
#poster or images

posters=soup.find_all('td',class_="posterColumn")
poster=[]
for i in range (0,len(posters)):
    poster.append(posters[i].src)
```

```python
#links

links=soup.find_all('td',class_='titleColumn')

link=[]
for i in range(0,len(links)):
    link.append(links[i].href)
```
