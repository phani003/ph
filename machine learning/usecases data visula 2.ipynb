```python
#creat bar plot
plt.figure(figsize=(15,15))
plt.bar(data['model'],data['cyl'])
plt.xlabel('Frequency of cars',size=15)
plt.ylabel(' Number of cylinders',size=15)
plt.xticks(rotation=60)
plt.show()
```

```python
df=data.isnull().sum().reset_index()
df.columns
```

```python
df.rename(columns={0:'na'},inplace=True)
```

```python
plt.bar(df['index'],df['na'],color='red')
plt.ylabel('count')
plt.xlabel('columns')
plt.xticks(rotation=60)
plt.show()
```
